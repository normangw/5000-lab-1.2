```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Plot
plt.figure(figsize=(6,4))
plt.plot(x, y, label="sin(x)", color="blue")
plt.title("Sample Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```
